```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("netflix.csv")
df
```

```python
for column in df.select_dtypes(include=['number']).columns:
  plt.figure(figsize=(12, 6))

  # Box plot
  plt.subplot(1, 2, 1)
  sns.boxplot(x=df[column])
  plt.title(f'Box Plot of {column}')

  # Histogram
  plt.subplot(1, 2, 2)
  plt.hist(df[column], bins=20)
  plt.title(f'Distribution of {column}')

  plt.show()
```

```python
Q1 = df['release_year'].quantile(0.25)
Q3 = df['release_year'].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['release_year'] >= lower_bound) & (df['release_year'] <= upper_bound)]

print(f"Number of rows before removing outliers: {df.shape[0]}")
print(f"Number of rows after removing outliers: {df_no_outliers.shape[0]}")
```

```python
plt.figure(figsize=(12, 6))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(x=df_no_outliers['release_year'])
plt.title(f'Box Plot of release_year (No Outliers)')

# Histogram
plt.subplot(1, 2, 2)
plt.hist(df_no_outliers['release_year'], bins=20)
plt.title(f'Distribution of release_year (No Outliers)')

plt.show()
```

```python
from scipy import stats

df['release_year_zscore'] = np.abs(stats.zscore(df['release_year']))

df_no_outliers_zscore = df[df['release_year_zscore'] <= 3]

df_no_outliers_zscore = df_no_outliers_zscore.drop('release_year_zscore', axis=1)

print(f"Number of rows before removing outliers: {df.shape[0]}")
print(f"Number of rows after removing outliers: {df_no_outliers_zscore.shape[0]}")

plt.figure(figsize=(12, 6))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(x=df_no_outliers_zscore['release_year'])
plt.title(f'Box Plot of release_year (No Outliers)')

# Histogram
plt.subplot(1, 2, 2)
plt.hist(df_no_outliers_zscore['release_year'], bins=20)
plt.title(f'Distribution of release_year (No Outliers)')

plt.show()
```

```python
# Select only numerical features
numerical_features = df.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_features.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
